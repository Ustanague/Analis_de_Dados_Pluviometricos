```python
import pandas as pd
from datetime import datetime, timedelta
import shutil
from os import remove
```

```python
#Importação das tabelas:
mes_fev = pd.read_fwf(r'D:\Estudo\Oficina de Pesquisa\Dados Brutos\Anchieta\anchieta_199702_Plv.txt')
mes_mar = pd.read_fwf(r'D:\Estudo\Oficina de Pesquisa\Dados Brutos\Anchieta\anchieta_199703_Plv.txt')
```

```python
mes_fev
```

```python
#Concatenando as Tabelas:
mes_concat = pd.concat([mes_fev, mes_mar], ignore_index = True)
mes_concat
```

```python
df_remove = mes_concat.loc[(mes_concat['Relatório Pluviométrico da Estação: Anchieta'] == "Dados normalizados para o Horário Brasileiro de Verão (HBV)") |
                      (mes_concat['Relatório Pluviométrico da Estação: Anchieta'] == "Dia         Hora      HBV   15 min   01 h   04 h   24 h   96 h")|
                      (mes_concat['Relatório Pluviométrico da Estação: Anchieta'] == "Relatório Pluviométrico da Estação: Anchieta")]
mes_concat = mes_concat.drop(df_remove.index)
```
