### Reimportando os dados

```python
import pandas as pd
from datetime import datetime, timedelta
import shutil
from os import remove
```

```python
colunas = ['Dia', 'Hora','HBV','15 min','01 h','04 h','24 h','96 h']
anchi_fev_1999 = pd.read_fwf(r'D:\Estudo\Oficina de Pesquisa\Teste\anchi_fev_1999.txt', names = colunas, sep ="/t")
```

```python
anchi_fev_1999
```

#  Criando Fevereiro 31 dias

```python
hora = timedelta(hours=1)

for i in range(0,anchi_fev_1999.shape[0]):
    if anchi_fev_1999.HBV[i] == "HBV":
        data_hbv = datetime.strptime(anchi_fev_1999.Dia[i]+" "+anchi_fev_1999.Hora[i],"%d/%m/%Y %H:%M:%S")
        correcao = data_hbv - hora
        data_corrigida = datetime.strftime(correcao,"%d/%m/%Y %H:%M:%S")
        dia, horario = data_corrigida.split()
        anchi_fev_1999.iat[i,0] = dia
        anchi_fev_1999.iat[i,1] = horario
```

```python
# Efetivamente 
anchi_fev31_1999 = anchi_fev_1999.loc[0:2979]
```
