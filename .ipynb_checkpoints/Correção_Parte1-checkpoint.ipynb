```python
import pandas as pd
from datetime import datetime, timedelta
import shutil
from os import remove
```

```python
#Importação das tabelas:
anchi_1999_fev = pd.read_fwf(r'D:\Estudo\Oficina de Pesquisa\Dados Brutos\Anchieta\anchieta_199902_Plv.txt')
anchi_1999_mar = pd.read_fwf(r'D:\Estudo\Oficina de Pesquisa\Dados Brutos\Anchieta\anchieta_199903_Plv.txt')
```

```python
anchi_1999_fev
```

```python
#Concatenando as Tabelas:
anchi_fev_1999 = pd.concat([anchi_1999_fev, anchi_1999_mar], ignore_index = True)
anchi_fev_1999
```

```python
df_remove = anchi_fev_1999.loc[(anchi_fev_1999['Relatório Pluviométrico da Estação: Anchieta'] == "Dados normalizados para o Horário Brasileiro de Verão (HBV)") |
                      (anchi_fev_1999['Relatório Pluviométrico da Estação: Anchieta'] == "Dia         Hora      HBV   15 min   01 h   04 h   24 h   96 h")|
                      (anchi_fev_1999['Relatório Pluviométrico da Estação: Anchieta'] == "Relatório Pluviométrico da Estação: Anchieta")]
anchi_fev_1999 = anchi_fev_1999.drop(df_remove.index)
```

```python
anchi_fev_1999
```

```python
anchi_fev_1999.to_csv(r'D:\Estudo\Oficina de Pesquisa\Teste\anchi_fev_1999.csv', index = False, sep = "\t")
shutil.copy(r'D:\Estudo\Oficina de Pesquisa\Teste\anchi_fev_1999.csv',r'D:\Estudo\Oficina de Pesquisa\Teste\anchi_fev_1999.txt')
remove(r'D:\Estudo\Oficina de Pesquisa\Teste\anchi_fev_1999.csv')
```
