```python
import pandas as pd
```

```python
Fev31_15min = []
FevNorm_15min = []
anos = ['1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013',
        '2014','2015','2016','2017','2018','2019','2020','2021','2022']
```

```python
pasta = input('Digite a Estação: ')
diretorio = input('Digite o diretório: ')
if diretorio == '':
    diretorio = r'D:\Estudo\Oficina de Pesquisa\Dados Tradados'
```

```python
#Tratando as entradas:
tamanho = pasta.split()
if len(tamanho) != 1:
    elemento = ('_'.join(tamanho)).lower()
else:
    elemento = pasta.lower()
```

```python
#Fevereiro de 31 Dias:
for ano in anos:
    arquivo = pd.read_csv(diretorio+'\\'+pasta+'\\'+elemento+'_fev31_'+ano+'.csv', sep=',')
    arquivo[arquivo['15 min'] == 'ND'] = 0 
    arquivo[['15 min']] = arquivo[['15 min']].apply(pd.to_numeric)
    soma = arquivo['15 min'].sum()
    Fev31_15min.append(soma)
```

```python
#Fevereiro Normal:
for ano in anos:
    arquivo = pd.read_csv(diretorio+'\\'+pasta+'\\'+elemento+'_Normal_'+ano+'.csv', sep=',')
    arquivo[arquivo['15 min'] == 'ND'] = 0 
    arquivo[['15 min']] = arquivo[['15 min']].apply(pd.to_numeric)
    soma = arquivo['15 min'].sum()
    FevNorm_15min.append(soma)
```

```python
lista_de_tuplas = list(zip(anos,Fev31_15min,FevNorm_15min))
```

```python
df = pd.DataFrame(lista_de_tuplas, columns=['Anos','Fev31','FevNormal'])
df.to_csv(diretorio+'\\'+elemento+'_Final.csv', index = False)
```
