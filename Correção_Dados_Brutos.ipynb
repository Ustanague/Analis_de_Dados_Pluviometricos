```python
import pandas as pd
from datetime import datetime, timedelta
import shutil
from os import remove, mkdir
```

```python
#Importação das tabelas:
ano_bissexto = [2000, 2004, 2008, 2012, 2016, 2020]
pasta = input('Digite a Estação: ')
#arquivo = input('Digite o arquivo: ')
ano = input('Digite o ano: ')
diretorio = 'D:\Estudo\Oficina de Pesquisa\Dados Brutos'
```

```python
#Tratando as entradas:
tamanho = pasta.split()
print(tamanho)
print(len(tamanho))
if len(tamanho) != 1:
    arquivo = ('_'.join(tamanho)).lower()
else:
    arquivo = pasta.lower()
print(arquivo)
```

```python
mes_fev = pd.read_fwf(diretorio+'\\'+pasta+'\\'+arquivo+'_'+ano+'02_Plv.txt',infer_datetime_format=True)
mes_mar = pd.read_fwf(diretorio+'\\'+pasta+'\\'+arquivo+'_'+ano+'03_Plv.txt',infer_datetime_format=True)
```

```python
#Concatenando as Tabelas:
mes_concat = pd.concat([mes_fev, mes_mar], ignore_index = False)
mes_concat
```

```python
#Valores a serem retirados:
df_remove = mes_concat.loc[(mes_concat['Relatório Pluviométrico da Estação: '+pasta] == "Dados normalizados para o Horário Brasileiro de Verão (HBV)") |
                      (mes_concat['Relatório Pluviométrico da Estação: '+pasta] == "Dia         Hora      HBV   15 min   01 h   04 h   24 h   96 h")|
                      (mes_concat['Relatório Pluviométrico da Estação: '+pasta] == "Relatório Pluviométrico da Estação: "+pasta)]
#Alteração nas tabelas:
mes_concat = mes_concat.drop(df_remove.index)
```

```python
# deleting a line
# based on the position
  
# opening the file in
# reading mode

with open(diretorio_tratado+'/'+pasta_saida+'/'+arquivo+'_'+ano+'.txt', 'r') as fr:
    # reading line by line
    lines = fr.readlines()
      
    # pointer for position
    ptr = 1
  
    # opening in writing mode
    with open(diretorio_tratado+'/'+pasta_saida+'/'+arquivo+'_'+ano+'.txt', 'w') as fw:
        for line in lines:
            
            # we want to remove 5th line
            if ptr != 1:
                fw.write(line)
            ptr += 1
```

```python
colunas = ['Dia', 'Hora','HBV','15 min','01 h','04 h','24 h','96 h']
meses = pd.read_fwf(diretorio_tratado+'\\'+pasta+'\\'+arquivo+'_'+ano+'.txt',
                    colspecs = [(0,10),(12,20),(22,25),(28,34),(35,41),(42,48),(49,55),(56,62)],
                    names = colunas)
meses
```

```python
hora = timedelta(hours=1)

for i in range(0,meses.shape[0]):
    if meses.HBV[i] == "HBV":
        data_hbv = datetime.strptime(meses.Dia[i]+" "+meses.Hora[i],"%d/%m/%Y %H:%M:%S")
        correcao = data_hbv - hora
        data_corrigida = datetime.strftime(correcao,"%d/%m/%Y %H:%M:%S")
        dia, horario = data_corrigida.split()
        meses.iat[i,0] = dia
        meses.iat[i,1] = horario
```

```python
#Preparando a condição:
ano_num = int(ano)
l_ano = [ano_num]
conj_ano = set(l_ano)
conj_bissexto = set(ano_bissexto)
```

```python
#Condição:
if (conj_ano.intersection(conj_bissexto)) == conj_ano:
    print('É bissexto:')
    #meses[['Dia']] = meses[['Dia']].apply(pd.to_datetime)
    meses['Dia'] = pd.to_datetime(meses['Dia'], dayfirst= True, infer_datetime_format=True)
    meses_id = meses.set_index('Dia')
    fev31 = meses_id.loc[: ano+"-03-02"]
    fevNormal = meses_id.loc[: ano+"-02-29"]
else:
    print('Não é bissexto')
    #meses[['Dia']] = meses[['Dia']].apply(pd.to_datetime)
    meses['Dia'] = pd.to_datetime(meses['Dia'], dayfirst= True, infer_datetime_format=True)
    meses_id = meses.set_index('Dia')
    fev31 = meses_id.loc[: ano+"-03-03"]
    fevNormal = meses_id.loc[: ano+"-02-28"]
```
