### Reimportando os dados

```python
import pandas as pd
from datetime import datetime, timedelta
import shutil
from os import remove
```

```python
colunas = ['Dia', 'Hora','HBV','15 min','01 h','04 h','24 h','96 h']
meses = pd.read_fwf(r'D:\Estudo\Oficina de Pesquisa\Teste\anchi_fev_1997.txt', names = colunas, sep ="/t")
```

```python
meses
```

#  Criando Fevereiro 31 dias

```python
hora = timedelta(hours=1)

for i in range(0,meses.shape[0]):
    if meses.HBV[i] == "HBV":
        data_hbv = datetime.strptime(meses.Dia[i]+" "+meses.Hora[i],"%d/%m/%Y %H:%M:%S")
        correcao = data_hbv - hora
        data_corrigida = datetime.strftime(correcao,"%d/%m/%Y %H:%M:%S")
        dia, horario = data_corrigida.split()
        meses.iat[i,0] = dia
        meses.iat[i,1] = horario
```
